```python
import pandas as pd
import numpy as np
import os
import warnings
warnings.filterwarnings('ignore')

# ── Chemins (adapter si besoin) ──────────────────────────────────
DATA_DIR   = "."        # dossier contenant les CSV sources
OUTPUT_DIR = "./output" # dossier de sortie
os.makedirs(OUTPUT_DIR, exist_ok=True)

pd.set_option('display.max_columns', None)
pd.set_option('display.max_colwidth', 60)
print('✅ Prêt')
```

```python
def transform_co2(df):
    df = df.copy()
    # Nettoyage des noms de colonnes
    df.columns = [
        'country', 'emissions_train_g_km', 'emissions_car_g_km',
        'emissions_plane_g_km', 'co2_saved_vs_car_g_km',
        'co2_saved_vs_plane_g_km', 'reduction_vs_car_pct'
    ]
    # Standardisation du nom de pays
    df['country'] = df['country'].str.strip()
    # S'assurer que les colonnes numériques sont bien float
    num_cols = df.columns[1:]
    df[num_cols] = df[num_cols].apply(pd.to_numeric, errors='coerce')
    print(f'✅ CO2 transformé : {df.shape}')
    return df

df_co2_clean = transform_co2(df_co2)
display(df_co2_clean.head())
```

```python
# Mapping pays → (code ISO, label anglais, label CO2)
COUNTRY_MAP = {
    'france'      : ('FR', 'France',      'France'),
    'germany'     : ('DE', 'Germany',     'Allemagne'),
    'italy'       : ('IT', 'Italy',       'Italie'),
    'spain'       : ('ES', 'Spain',       'Espagne'),
    'switzerland' : ('CH', 'Switzerland', 'Suisse'),
}

# Colonnes standardisées communes aux 5 fichiers nationaux
NATIONAL_COLS = [
    'year', 'origin_station', 'destination_station',
    'origin_city', 'destination_city',
    'origin_region', 'destination_region',
    'passengers_millions', 'service_type', 'source'
]

def transform_od_national(df, country_key):
    """Nettoyage générique pour les flux OD nationaux."""
    df = df.copy()
    iso, label, _ = COUNTRY_MAP[country_key]

    df.columns = NATIONAL_COLS

    # Métadonnées pays
    df['country']     = label
    df['country_iso'] = iso
    df['flow_type']   = 'domestic'

    # Cast types
    df['year']               = pd.to_numeric(df['year'], errors='coerce').astype('Int64')
    df['passengers_millions'] = pd.to_numeric(df['passengers_millions'], errors='coerce')

    # Nettoyage textes
    text_cols = ['origin_station', 'destination_station',
                 'origin_city', 'destination_city',
                 'origin_region', 'destination_region']
    for col in text_cols:
        df[col] = df[col].str.strip()

    # Suppression doublons
    before = len(df)
    df.drop_duplicates(subset=['year', 'origin_station', 'destination_station'], inplace=True)
    removed = before - len(df)
    if removed:
        print(f'  ⚠️  {removed} doublon(s) supprimé(s) dans {label}')

    # Suppression lignes sans passagers
    df.dropna(subset=['passengers_millions'], inplace=True)

    print(f'✅ {label:<15} transformé : {df.shape[0]} lignes')
    return df

df_france_clean  = transform_od_national(df_france,  'france')
df_germany_clean = transform_od_national(df_germany, 'germany')
df_italy_clean   = transform_od_national(df_italy,   'italy')
df_spain_clean   = transform_od_national(df_spain,   'spain')
df_switz_clean   = transform_od_national(df_switz,   'switzerland')
```

```python
def transform_crossborder(df):
    df = df.copy()
    df.columns = [
        'year',
        'origin_country_iso', 'origin_station', 'origin_city', 'origin_region',
        'destination_country_iso', 'destination_station', 'destination_city', 'destination_region',
        'passengers_millions', 'service_type', 'source'
    ]
    df['flow_type'] = 'cross_border'

    df['year']                = pd.to_numeric(df['year'], errors='coerce').astype('Int64')
    df['passengers_millions'] = pd.to_numeric(df['passengers_millions'], errors='coerce')

    text_cols = ['origin_station', 'destination_station',
                 'origin_city', 'destination_city',
                 'origin_region', 'destination_region']
    for col in text_cols:
        df[col] = df[col].str.strip()

    # Normaliser les noms de pays ISO (ex: 'Italy' → 'IT')
    country_to_iso = {
        'France': 'FR', 'Germany': 'DE', 'Italy': 'IT',
        'Spain': 'ES', 'Switzerland': 'CH'
    }
    df['origin_country_iso']      = df['origin_country_iso'].map(country_to_iso).fillna(df['origin_country_iso'])
    df['destination_country_iso'] = df['destination_country_iso'].map(country_to_iso).fillna(df['destination_country_iso'])

    df.dropna(subset=['passengers_millions'], inplace=True)
    print(f'✅ Cross-border transformé : {df.shape[0]} lignes')
    return df

df_crossbord_clean = transform_crossborder(df_crossbord)
display(df_crossbord_clean.head())
```

```python
# Schéma commun minimal pour le concat
COMMON_COLS = [
    'year', 'origin_station', 'destination_station',
    'origin_city', 'destination_city',
    'origin_region', 'destination_region',
    'passengers_millions', 'service_type', 'flow_type', 'source'
]

# Flux nationaux : on ajoute origin/destination_country_iso identiques
national_frames = []
for df, iso in [
    (df_france_clean,  'FR'),
    (df_germany_clean, 'DE'),
    (df_italy_clean,   'IT'),
    (df_spain_clean,   'ES'),
    (df_switz_clean,   'CH'),
]:
    tmp = df[COMMON_COLS].copy()
    tmp['origin_country_iso']      = iso
    tmp['destination_country_iso'] = iso
    national_frames.append(tmp)

# Flux cross-border
cb = df_crossbord_clean[COMMON_COLS + ['origin_country_iso', 'destination_country_iso']].copy()

# Concatenation
df_all_od = pd.concat(national_frames + [cb], ignore_index=True)

print(f'✅ Table OD consolidée : {df_all_od.shape[0]:,} lignes | {df_all_od.shape[1]} colonnes')
display(df_all_od.sample(5, random_state=42))
```

```python
# Mapping ISO → nom dans la table CO2 (noms en français)
ISO_TO_CO2_COUNTRY = {
    'FR': 'France',
    'DE': 'Allemagne',
    'IT': 'Italie',
    'ES': 'Espagne',
    'CH': 'Suisse',
}

df_all_od['co2_country_key'] = df_all_od['origin_country_iso'].map(ISO_TO_CO2_COUNTRY)

df_enriched = df_all_od.merge(
    df_co2_clean[['country', 'emissions_train_g_km', 'emissions_car_g_km',
                  'co2_saved_vs_car_g_km', 'reduction_vs_car_pct']],
    left_on='co2_country_key',
    right_on='country',
    how='left'
).drop(columns=['country', 'co2_country_key'])

print(f'✅ Table enrichie : {df_enriched.shape}')
print(f"   Taux de match CO2 : {df_enriched['emissions_train_g_km'].notna().mean():.1%}")
display(df_enriched.head())
```

```python
print('=== Répartition par pays et type de flux ===')
display(
    df_enriched.groupby(['origin_country_iso', 'flow_type'])
               .agg(
                   nb_routes          = ('passengers_millions', 'count'),
                   total_passengers_M = ('passengers_millions', 'sum'),
               )
               .round(2)
               .reset_index()
               .sort_values('total_passengers_M', ascending=False)
)
```

```python
print('=== Top 10 des routes les plus fréquentées ===')
top10 = (
    df_enriched
    .groupby(['origin_country_iso', 'origin_city', 'destination_city'])
    ['passengers_millions'].sum()
    .reset_index()
    .sort_values('passengers_millions', ascending=False)
    .head(10)
)
display(top10)
```

```python
print('=== Évolution totale par année ===')
display(
    df_enriched.groupby('year')
               .agg(
                   nb_routes          = ('passengers_millions', 'count'),
                   total_passengers_M = ('passengers_millions', 'sum'),
               )
               .round(2)
)
```

```python
print('=== Comparatif CO₂ par pays (train vs voiture) ===')
display(df_co2_clean[['country', 'emissions_train_g_km', 'emissions_car_g_km',
                       'reduction_vs_car_pct']].sort_values('emissions_train_g_km'))
```

```python
# ── Table 1 : flux OD enrichis (table principale) ────────────────
out_od = os.path.join(OUTPUT_DIR, 'od_flows_enriched.csv')
df_enriched.to_csv(out_od, index=False, sep=';', encoding='utf-8-sig')
print(f'💾 od_flows_enriched.csv     → {len(df_enriched):,} lignes')

# ── Table 2 : CO2 nettoyé ─────────────────────────────────────────
out_co2 = os.path.join(OUTPUT_DIR, 'co2_clean.csv')
df_co2_clean.to_csv(out_co2, index=False, sep=';', encoding='utf-8-sig')
print(f'💾 co2_clean.csv             → {len(df_co2_clean):,} lignes')

# ── Table 3 : agrégat par pays/année/type ─────────────────────────
df_agg = (
    df_enriched
    .groupby(['year', 'origin_country_iso', 'flow_type'])
    .agg(
        nb_routes          = ('passengers_millions', 'count'),
        total_passengers_M = ('passengers_millions', 'sum'),
        avg_passengers_M   = ('passengers_millions', 'mean'),
        max_route_M        = ('passengers_millions', 'max'),
    )
    .round(3)
    .reset_index()
)
out_agg = os.path.join(OUTPUT_DIR, 'od_aggregated.csv')
df_agg.to_csv(out_agg, index=False, sep=';', encoding='utf-8-sig')
print(f'💾 od_aggregated.csv         → {len(df_agg):,} lignes')

print(f'\n✅ Export terminé — dossier : {os.path.abspath(OUTPUT_DIR)}')
```

```python
import pandas as pd
import sqlite3
import os

# ── Chemins (adapter si besoin) ──────────────────────────────────
OUTPUT_DIR = "./output"                        # dossier des CSV produits par l'ETL
DB_PATH    = "./rail_europe.db"                # fichier SQLite à créer

print(f'📁 Dossier CSV  : {os.path.abspath(OUTPUT_DIR)}')
print(f'🗄️  Base SQLite  : {os.path.abspath(DB_PATH)}')
print('✅ Prêt')
```

```python
def read_csv(filename):
    path = os.path.join(OUTPUT_DIR, filename)
    df = pd.read_csv(path, sep=';', encoding='utf-8-sig')
    print(f'📂  {filename:<35}  {df.shape[0]:>5} lignes | {df.shape[1]} colonnes')
    return df

df_od  = read_csv('od_flows_enriched.csv')
df_co2 = read_csv('co2_clean.csv')
df_agg = read_csv('od_aggregated.csv')
print('\n✅ CSV chargés en mémoire.')
```

```python
# Connexion (crée le fichier s'il n'existe pas)
con = sqlite3.connect(DB_PATH)
print(f'🔌 Connexion ouverte → {DB_PATH}')

# ── Schémas DDL explicites ────────────────────────────────────────
DDL = """
-- Table principale : flux OD enrichis
CREATE TABLE IF NOT EXISTS od_flows_enriched (
    id                       INTEGER PRIMARY KEY AUTOINCREMENT,
    year                     INTEGER,
    origin_station           TEXT,
    destination_station      TEXT,
    origin_city              TEXT,
    destination_city         TEXT,
    origin_region            TEXT,
    destination_region       TEXT,
    passengers_millions      REAL,
    service_type             TEXT,
    flow_type                TEXT,
    source                   TEXT,
    origin_country_iso       TEXT,
    destination_country_iso  TEXT,
    emissions_train_g_km     REAL,
    emissions_car_g_km       REAL,
    co2_saved_vs_car_g_km    REAL,
    reduction_vs_car_pct     REAL
);

-- Table CO2 par pays
CREATE TABLE IF NOT EXISTS co2_clean (
    id                       INTEGER PRIMARY KEY AUTOINCREMENT,
    country                  TEXT,
    emissions_train_g_km     REAL,
    emissions_car_g_km       REAL,
    emissions_plane_g_km     REAL,
    co2_saved_vs_car_g_km    REAL,
    co2_saved_vs_plane_g_km  REAL,
    reduction_vs_car_pct     REAL
);

-- Table agrégat par pays/année/type
CREATE TABLE IF NOT EXISTS od_aggregated (
    id                     INTEGER PRIMARY KEY AUTOINCREMENT,
    year                   INTEGER,
    origin_country_iso     TEXT,
    flow_type              TEXT,
    nb_routes              INTEGER,
    total_passengers_M     REAL,
    avg_passengers_M       REAL,
    max_route_M            REAL
);
"""

con.executescript(DDL)
con.commit()
print('✅ Tables créées (ou déjà existantes).')
```

```python
# ── Insertion des données ─────────────────────────────────────────
# if_exists='replace' : vide et recrée la table à chaque exécution
# Utiliser 'append' si tu veux cumuler des runs

tables = {
    'od_flows_enriched' : df_od,
    'co2_clean'         : df_co2,
    'od_aggregated'     : df_agg,
}

for table_name, df in tables.items():
    df.to_sql(table_name, con, if_exists='replace', index=False)
    # Vérification du nombre de lignes insérées
    count = con.execute(f'SELECT COUNT(*) FROM {table_name}').fetchone()[0]
    print(f'✅ {table_name:<25} → {count:>5} lignes insérées')

con.commit()
print('\n💾 Toutes les tables sont chargées.')
```

```python
INDEXES = """
CREATE INDEX IF NOT EXISTS idx_od_country_year
    ON od_flows_enriched (origin_country_iso, year);

CREATE INDEX IF NOT EXISTS idx_od_flow_type
    ON od_flows_enriched (flow_type);

CREATE INDEX IF NOT EXISTS idx_od_cities
    ON od_flows_enriched (origin_city, destination_city);

CREATE INDEX IF NOT EXISTS idx_agg_country_year
    ON od_aggregated (origin_country_iso, year);
"""

con.executescript(INDEXES)
con.commit()
print('✅ Index créés.')
```

```python
# Vue d'ensemble de la base
tables_in_db = pd.read_sql("SELECT name FROM sqlite_master WHERE type='table'", con)
print('Tables dans la base :', tables_in_db['name'].tolist())

print('\n=== Taille du fichier SQLite ===')
size_kb = os.path.getsize(DB_PATH) / 1024
print(f'📦 {DB_PATH} : {size_kb:.1f} Ko')
```

```python
# Requête 1 : Top 10 routes les plus fréquentées
q1 = """
SELECT
    origin_country_iso,
    origin_city,
    destination_city,
    ROUND(SUM(passengers_millions), 2) AS total_pax_M
FROM od_flows_enriched
GROUP BY origin_country_iso, origin_city, destination_city
ORDER BY total_pax_M DESC
LIMIT 10;
"""
print('=== Top 10 routes ===')
display(pd.read_sql(q1, con))
```

```python
# Requête 2 : Passagers totaux par pays et type de flux
q2 = """
SELECT
    origin_country_iso AS pays,
    flow_type,
    COUNT(*)                              AS nb_routes,
    ROUND(SUM(passengers_millions), 2)    AS total_pax_M
FROM od_flows_enriched
GROUP BY pays, flow_type
ORDER BY total_pax_M DESC;
"""
print('=== Passagers par pays / type ===')
display(pd.read_sql(q2, con))
```

```python
# Requête 3 : Émissions CO2 et économies par pays
q3 = """
SELECT
    country,
    emissions_train_g_km,
    emissions_car_g_km,
    co2_saved_vs_car_g_km,
    reduction_vs_car_pct
FROM co2_clean
ORDER BY emissions_train_g_km ASC;
"""
print('=== CO₂ — train vs voiture ===')
display(pd.read_sql(q3, con))
```

```python
# Requête 4 : Évolution annuelle (jointure od_flows + co2)
q4 = """
SELECT
    o.year,
    o.origin_country_iso AS pays,
    ROUND(SUM(o.passengers_millions), 2)          AS total_pax_M,
    MAX(o.emissions_train_g_km)                   AS co2_train_g_km,
    ROUND(SUM(o.passengers_millions)
          * MAX(o.co2_saved_vs_car_g_km), 0)       AS co2_economise_total_g
FROM od_flows_enriched o
WHERE o.flow_type = 'domestic'
GROUP BY o.year, o.origin_country_iso
ORDER BY o.year, pays;
"""
print('=== Évolution annuelle + CO₂ économisé (domestic) ===')
display(pd.read_sql(q4, con))
```

```python
con.close()
print(f'🔒 Connexion fermée.')
print(f'\n✅ Base de données prête : {os.path.abspath(DB_PATH)}')
print('   → Ouvrir avec DB Browser for SQLite, DBeaver, ou tout autre client SQLite.')
```

```python
import sqlite3
import pandas as pd
import os

# Modifier ce chemin selon votre environnement
DOSSIER = "."  # Dossier contenant les CSV et la BD
DB_PATH = os.path.join(DOSSIER, "rail_europe.db")

# Mapping fichier CSV -> nom de table
FICHIERS_CSV = {
    "agencies.csv":          "agencies",
    "calendrier.csv":        "calendrier",
    "calendrier_dates.csv":  "calendrier_dates",
    "classes.csv":           "classes",
    "routes.csv":            "routes",
    "stops.csv":             "stops",
    "translations.csv":      "translations",
    "trip_stop.csv":         "trip_stop",
    "trips.csv":             "trips",
    "view_ontd_cities.csv":  "view_ontd_cities",
    "view_ontd_details.csv": "view_ontd_details",
    "view_ontd_list.csv":    "view_ontd_list",
    "view_ontd_map.csv":     "view_ontd_map",
}

print(f"Base de données : {os.path.abspath(DB_PATH)}")
print(f"Dossier source  : {os.path.abspath(DOSSIER)}")
print(f"Fichiers à importer : {len(FICHIERS_CSV)}")
```

```python
# Vérifier que tous les fichiers sont présents
for fichier in FICHIERS_CSV:
    chemin = os.path.join(DOSSIER, fichier)
    if os.path.exists(chemin):
        taille = os.path.getsize(chemin)
        print(f"  ✓ {fichier:30s} ({taille:>10,} octets)")
    else:
        print(f"  ✗ {fichier:30s} INTROUVABLE")

# Vérifier la BD
if os.path.exists(DB_PATH):
    print(f"\n  ✓ Base de données trouvée ({os.path.getsize(DB_PATH):,} octets)")
else:
    print(f"\n  ℹ Base de données sera créée : {DB_PATH}")
```

```python
# Aperçu rapide de chaque fichier
for fichier, table in FICHIERS_CSV.items():
    chemin = os.path.join(DOSSIER, fichier)
    if not os.path.exists(chemin):
        continue
    df = pd.read_csv(chemin, nrows=3, encoding="utf-8")
    print(f"\n{'='*60}")
    print(f"{fichier} → table '{table}' | {len(df.columns)} colonnes")
    print(f"Colonnes : {list(df.columns)}")
    display(df.head(3))
```

```python
conn = sqlite3.connect(DB_PATH)
resultats = []

for fichier, nom_table in FICHIERS_CSV.items():
    chemin = os.path.join(DOSSIER, fichier)
    
    if not os.path.exists(chemin):
        print(f"  ⚠ ABSENT : {fichier}")
        resultats.append({"fichier": fichier, "table": nom_table, "statut": "ABSENT", "lignes": 0})
        continue
    
    try:
        # Lecture du CSV
        df = pd.read_csv(chemin, encoding="utf-8", low_memory=False)
        df.columns = df.columns.str.strip()
        
        # Chargement dans SQLite (remplace la table si elle existe)
        df.to_sql(nom_table, conn, if_exists="replace", index=False)
        
        # Vérification
        count = pd.read_sql(f'SELECT COUNT(*) as n FROM "{nom_table}"', conn).iloc[0, 0]
        
        print(f"  ✓ {fichier:30s} → {nom_table:25s} | {count:>7,} lignes | {len(df.columns)} col")
        resultats.append({
            "fichier": fichier, "table": nom_table,
            "statut": "OK", "lignes": count, "colonnes": len(df.columns)
        })
    except Exception as e:
        print(f"  ✗ ERREUR sur {fichier}: {e}")
        resultats.append({"fichier": fichier, "table": nom_table, "statut": f"ERREUR", "lignes": 0})

conn.commit()
print(f"\n{'='*60}")
print(f"Import terminé : {sum(1 for r in resultats if r['statut']=='OK')}/{len(FICHIERS_CSV)} fichiers OK")
print(f"Total : {sum(r['lignes'] for r in resultats):,} lignes importées")
```

```python
# Index pour optimiser les requêtes fréquentes
INDEX_SQL = [
    # Tables GTFS principales
    'CREATE INDEX IF NOT EXISTS idx_routes_agency ON routes(agency_id);',
    'CREATE INDEX IF NOT EXISTS idx_trips_route ON trips(route_id);',
    'CREATE INDEX IF NOT EXISTS idx_trip_stop_trip ON trip_stop(trip_id);',
    'CREATE INDEX IF NOT EXISTS idx_trip_stop_stop ON trip_stop(stop_id);',
    'CREATE INDEX IF NOT EXISTS idx_stops_name ON stops(stop_name);',
    'CREATE INDEX IF NOT EXISTS idx_calendrier_service ON calendrier(service_id);',
    'CREATE INDEX IF NOT EXISTS idx_calendrier_dates_service ON calendrier_dates(service_id);',
    'CREATE INDEX IF NOT EXISTS idx_translations_table ON translations(table_name);',
]

cursor = conn.cursor()
for sql in INDEX_SQL:
    try:
        cursor.execute(sql)
        nom_index = sql.split('idx_')[1].split(' ')[0]
        print(f"  ✓ Index créé : idx_{nom_index}")
    except Exception as e:
        print(f"  ⚠ {sql[:50]}... → {e}")

conn.commit()
print("\nIndex créés avec succès.")
```

```python
# Liste de toutes les tables et leur contenu
tables_info = pd.read_sql("""
    SELECT name as table_name 
    FROM sqlite_master 
    WHERE type='table' 
    ORDER BY name
""", conn)

print(f"{'Table':30s} | {'Lignes':>10s} | {'Colonnes':>10s}")
print("-" * 60)

for table in tables_info['table_name']:
    count = pd.read_sql(f'SELECT COUNT(*) as n FROM "{table}"', conn).iloc[0, 0]
    cols = pd.read_sql(f'PRAGMA table_info("{table}")', conn)
    print(f"{table:30s} | {count:>10,} | {len(cols):>10}")

print(f"\nTotal : {len(tables_info)} tables dans la base de données")
```

```python
# Schéma détaillé de chaque table (colonnes + types)
for table in tables_info['table_name']:
    schema = pd.read_sql(f'PRAGMA table_info("{table}")', conn)
    print(f"\n--- {table} ---")
    for _, row in schema.iterrows():
        print(f"  {row['name']:30s} {row['type']}")
```

```python
# Test : requête exemple
print("Exemple de requête : premières lignes de chaque table\n")

for table in tables_info['table_name']:
    print(f"\n{'='*60}")
    print(f"Table : {table}")
    display(pd.read_sql(f'SELECT * FROM "{table}" LIMIT 3', conn))
```

```python
import sqlite3
import pandas as pd
import os

# Modifier ces chemins selon votre environnement
DB_PATH = "rail_europe.db"
CSV_PATH = "empreinte_carbone_trajets.csv"
NOM_TABLE = "empreinte_carbone_trajets"

print(f"Base de données : {os.path.abspath(DB_PATH)}")
print(f"Fichier CSV     : {os.path.abspath(CSV_PATH)}")
print(f"Table cible     : {NOM_TABLE}")
```

```python
df = pd.read_csv(CSV_PATH, encoding="utf-8")
df.columns = df.columns.str.strip()

print(f"{len(df)} lignes | {len(df.columns)} colonnes")
print(f"\nColonnes : {list(df.columns)}")
df.head(5)
```

```python
conn = sqlite3.connect(DB_PATH)

# Import dans la BD (remplace la table si elle existe déjà)
df.to_sql(NOM_TABLE, conn, if_exists="replace", index=False)

# Vérification
count = pd.read_sql(f'SELECT COUNT(*) as n FROM "{NOM_TABLE}"', conn).iloc[0, 0]
print(f"✓ Table '{NOM_TABLE}' créée : {count} lignes")
```

```python
cursor = conn.cursor()

INDEX = [
    f'CREATE INDEX IF NOT EXISTS idx_ect_transporteur ON "{NOM_TABLE}"(transporteur);',
    f'CREATE INDEX IF NOT EXISTS idx_ect_origine ON "{NOM_TABLE}"(origine);',
    f'CREATE INDEX IF NOT EXISTS idx_ect_destination ON "{NOM_TABLE}"(destination);',
    f'CREATE INDEX IF NOT EXISTS idx_ect_origine_uic ON "{NOM_TABLE}"(origine_uic);',
    f'CREATE INDEX IF NOT EXISTS idx_ect_destination_uic ON "{NOM_TABLE}"(destination_uic);',
]

for sql in INDEX:
    cursor.execute(sql)
    nom = sql.split('idx_')[1].split(' ')[0]
    print(f"  ✓ Index créé : idx_{nom}")

conn.commit()
print("\nIndex créés avec succès.")
```

```python
schema = pd.read_sql(f'PRAGMA table_info("{NOM_TABLE}")', conn)
print(f"Schéma de la table '{NOM_TABLE}' :\n")
for _, row in schema.iterrows():
    print(f"  {row['name']:55s} {row['type']}")
```
