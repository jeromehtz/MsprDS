```python
import pandas as pd
import sqlite3
import os

fichiers_a_charger={
    'co2_comparaison_europe.csv':'co2europe',
    'frequentation_gares_france.csv':'gare_france',
    'frequentation_gares_allemagne.csv':'gare_allemagne',
    'frequentation_gares_italie.csv':'gare_italie',
    'frequentation_gares_portugal.csv':'gare_portugal',
    'frequentation_gares_suisse.csv':'gare_suisse',
    'temps_trajet_trains_allemagne.csv':'trajet_allemagne',
    'temps_trajet_trains_france.csv':'trajet_france',
    'temps_trajet_trains_italie.csv':'trajet_italie',
    'temps_trajet_trains_portugal.csv':'trajet_portugal',
    'temps_trajet_trains_suisse.csv':'trajet_suisse',
    'liaisons_internationales_europe.csv':'trajet_pays'}


conn = sqlite3.connect('mobilite_ia.db')

for fichier, table in fichiers_a_charger.items():
    if os.path.exists(fichier):
        
        try:
           
            df = pd.read_csv(fichier, sep=';', encoding='utf-8-sig')
            
           
            df.to_sql(table, conn, if_exists='replace', index=False)
            print(f"✅ Table '{table}' chargée avec succès depuis {fichier}")
            
        except Exception as e:
            print(f"❌ Erreur lors du chargement de {fichier}: {e}")
    else:
        print(f"⚠️ Le fichier {fichier} est introuvable. Assurez-vous qu'il est dans le dossier.")

conn.close()
print("\nBase de données 'mobilite_ia.db' prête pour utilisation.")

    
    
```
